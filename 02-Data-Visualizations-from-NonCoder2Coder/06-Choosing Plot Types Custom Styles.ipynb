有时，不是轻而易举就能发现我们数据背后的联系，在这里，我们将绘图分为三个类型来理解：  

1.趋势图--表征一种变化的形式  
**sns.lineplot**  

2.关系图--有很多种方法来发现和理解不同变量之间的关系  
**柱状图sns.barplot**  
**热度图sns.headmap**  
**散点图sns.scatter**  
**回归直线sns.regplot**  
**多条回归直线sns.lmplot**  
**分类散点图sns.swarmplot**  

3.分布图--我们将分布可视化，验证那些我们期望看到的值  
**直方图sns.distplot**  
**密度图sns.kdeplot**  
**二维密度图sns.jointplot**
